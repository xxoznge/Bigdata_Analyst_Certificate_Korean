```python
# 데이터프레임객체.sort_values(by, axis=0, ascending=True, inplace=False, ...)

obj = {'Name' : ['dfs', 'sd', 'sdf', 'sdfsd'],
       'Sex' : ['f', 'm', 'f', 'f'],
       'Age' : [11, 22, 33, 44],
       'Score' : [100, 90, 80, 70]}

df = pd.DataFrame(obj)
```

```python
df.sort_values('Score')
```

```python
df.sort_values(['Age', 'Score'], ascending = [True, False]) #age는 오름차순, score는 내림차순
```

```python
# 데이터프레임객체.sort_index(axis=0, level=None, ascending=True, inplace=False, ...)

rownm = ['dfsdf', 'sdfsdf', 'sdfsdf', 'sdfsdf']
obj = {'Sex' : ['f', 'm', 'f', 'f'],
       'Age' : [11, 22, 33, 44],
       'Score' : [100, 90, 80, 70]}
df = pd.DataFrame(obj, rownm)
df
```

```python
print(df.sort_index()) # 행 오름차순
print(df.sort_index(axis=1)) # 열 오름차순
```

```python
# 데이터 순위
# 데이터프레임객체.rank(axis=1, method='average',ascending=True, ...)
```

```python
df.sort_values('Score', ascending = False, inplace = True)
df.head()
```

```python
# case1. 동점자 평균 순위
df['rank_avg1'] = df['Score'].rank()
```

```python
# case2. 동점자 평균 순위(가장 큰 값이 1위)
# Score가 90인 경우가 3개로, 3~5위의 평균인 4위
df['rank_avg1'] = df['Score'].rank(ascending = False)
```

```python
# case3. 동점자 가장 낮은 순위(가장 큰 값이 1위)
# Score가 90인 경우가 3개로, 3~5위 중 가장 낮은 5위
df['rank_min'] = df['Score'].rank(method='min', ascending=False)
```

```python
# case4. 동점자 가장 높은 순위(가장 큰 값이 1위)
# Score가 90인 경우가 3개로, 3~5위 중 가장 높은 5위
df['rank_max'] = df['Score'].rank(method='max', ascending=False)
```
