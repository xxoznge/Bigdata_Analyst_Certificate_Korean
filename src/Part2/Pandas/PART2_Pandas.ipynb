```python
# 1. Series : 일차원 배열 데이터인 값과 데이터의 레이블 배열인 인덱스로 구성 / 데이터 타입이 달라도 됨
# pandas.Series(object, index)

import pandas as pd
import numpy as np
```

```python
obj = ['이유리', '최민준', '호세헌']
pd.Series(obj)
```

```python
idx = ['A', 'B', 'C']
pd.Series(obj,idx)
```

```python
# Series의 인덱싱과 슬라이싱
```

```python
obj = [22,32,27,18,38,19]
idx = list('abcdef')
sr = pd.Series(obj, idx)
sr
```

```python
print(sr[0])
print(sr['e'])
print(sr[1:5])

idx = [0, 3, 4]
sr[idx]
ldl = ['a', 'e']
sr[ldl]
```

```python
# 통계 메소드
# describe() : 요약, count(), mean(), var(), std(), min(), max(), median(), quantile(), unique(), value_counts() : 유일값 빈도수, mode() : 최빈값
```

```python
sr = pd.Series(range(100))
```

```python
sr.describe()
```

```python
print(sr.count())
print(sr.mean())
print(sr.var())
print(sr.std())
print(sr.min())
print(sr.max())
print(sr.median())
print(sr.quantile())
print(sr.quantile(q = 0.25))  # 4분위수
print(sr.unique)
```

```python
sr = pd.Series(['a', 'b', 'c', 'd'])
```

```python
print(sr.describe())
print(sr.unique())
print(sr.value_counts())
print(sr.mode())
```

```python
# 2. DataFrame : 같은 크기의 열들을 모아놓은 구조적 2차원 객체 / 열, 행, 인덱스
# pandas.DataFrame(object, index, columns)
```

```python
obj = {'이름' : ['호세헌', '세헌'],
       '전공' : ['경영학과', '컴퓨터공학과'],
       '성별' : ['여', '남'],
       '나이' : [22, 25]}
df = pd.DataFrame(obj)
df
```

```python
obj = np.array([['호세헌', '경영학과', '여', 20],
                ['세헌', '컴퓨터공학과', '남', 25]])
df1 = pd.DataFrame(obj)
df1
```

```python
df2 = pd.DataFrame(obj,
                   columns = ['Name', 'Major', 'Sex', 'Age'],
                   index = ['A', 'B']
                   )
df2
```

```python
# 정보 확인
# values, index, columns, dtypes, shape, info(), head(n=5), tail(n=5)
```

```python
obj = {'Name' : ['dfs', 'sd', 'sdf', 'sdfsd'],
       'Sex' : ['f', 'm', 'f', 'f'],
       'Age' : [11, 22, 33, 44]}
df = pd.DataFrame(obj)
df
```

```python
print(df.values)
print(df.index)
print(df.columns)
print(df.dtypes)
print(df.info)
print(df.shape) # 행, 열 값
print(df.head(2))
print(df.tail(1))
```

```python
# 데이터프레임의 인덱싱과 슬라이싱
```

```python
obj = {'Name' : ['dfs', 'sd', 'sdf', 'sdfsd'],
       'Sex' : ['f', 'm', 'f', 'f'],
       'Age' : [11, 22, 33, 44]}
idx = list('abcd')
df = pd.DataFrame(obj, idx)
df
```

```python
print(df['Name'])
print(df.Name)

colnm = ['Name', 'Age']
df[colnm]
```

```python
print(df.iloc[1])
print(df.iloc[1:3])

idx = [1,3]
print(df.iloc[idx])
```

```python
rownm = ['b','d']
df.loc[rownm]
```

```python
df['Age']['a']
```

```python
df.iloc[1:3,1:]
```

```python
rownm = ['a', 'c']
colnm = ['Name', 'Age']
df.loc[rownm, colnm]
```
