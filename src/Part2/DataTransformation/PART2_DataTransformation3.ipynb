```python
# 1. 표준화 : 평균을 기준으로 얼마나 떨어져 있는지

from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# 표준 스케일러 생성
std_scale = StandardScaler()

# 표준화 데이터로 변환
ds = std_scale.fit_transform(data)
ds = pd.DataFrame(iris.data, columns=iris.feature_names)
ds.head()
```

```python
# 2. 정규화 : 0과 1 사이로

from sklearn.preprocessing import MinMaxScaler

iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

mm_scale = MinMaxScaler()

ds = mm_scale.fit_transform(data)
ds = pd.DataFrame(ds, columns = iris.feature_names)
ds.head()
```

```python
# 1. 주성분분석 : 상관성이 높은 변수들의 선형결합 / 주성분 기여율 : 주성분 설명력 / 85% 이상되면 거기까지

# USArrests 데이터로 주성분분석 수행 - 잘 요약해라

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv("USArrests_rownames.csv", index_col = 0)
df.head
```

```python
# 표준화하기
x = StandardScaler().fit_transform(df)

# PCA 객체 생성
pca = PCA(n_components = 4)

# 적합 후 변환 (배열 형태))
pca_arr = pca.fit_transform(x)

# 주성분에 따른 분산 기여율, 누적 기여율
result = {"분산 기여율" : pca.explained_variance_ratio_,
          "누적 기여율" : pca.explained_variance_ratio_.cumsum()}

# 결과를 데이터 프레임으로, 전치해서 컬럼명이 주성분으로 되게
result = pd.DataFrame(result,
                      index = ['PCA1', 'PCA2', 'PCA3', 'PCA4']).T
print(result)

# 주성분 분석 결과
pca = PCA(n_components = 2) # PCA 객체 생성
pca_arr = pca.fit_transform(x) # 적합 후 변환
df_pca = pd.DataFrame(pca_arr, columns = ['PCA1', 'PCA2']) # 데이터 프레임으로 변환
df_pca.head() # 결과 확인 5개
```
